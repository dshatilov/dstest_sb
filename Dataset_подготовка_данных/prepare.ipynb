```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
train_data = pd.read_csv('./train/X_train.txt',  engine='python', delim_whitespace=True, header=None)
test_data = pd.read_csv('./test/X_test.txt',  engine='python', delim_whitespace=True, header=None)
sub_train = pd.read_csv('./train/subject_train.txt',  engine='python', header=None)
sub_test = pd.read_csv('./test/subject_test.txt',  engine='python', header=None)
act_train = pd.read_csv('./train/y_train.txt',  engine='python', header=None)
act_test = pd.read_csv('./test/y_test.txt',  engine='python', header=None)
features = pd.read_csv('./features.txt', engine='python', header=None, delim_whitespace=True, names=['NN', 'feature'])
act_code = pd.read_csv('./activity_labels.txt',  engine='python', header=None, delim_whitespace=True, names=['Act_Code', 'kind'])
```

```python
subjects = sub_train.append(sub_test)
subjects.reset_index(drop=True);
activity = act_train.append(act_test)
activity.reset_index(drop=True);
data = train_data.append(test_data)
data.reset_index(drop=True);
data.insert(0,'Subject_Code', subjects)
data.insert(1, 'Act_Code', activity)

features.drop(columns=['NN'], inplace=True)
a = ['-mean()' in pr for pr in list(features.feature)] 
b = ['-std()' in pr for pr in list(features.feature)]
c = ['meanFreq' in pr for pr in list(features.feature)]
mask = [not (a[i] or b[i] and not c[i]) for i in range(len(a))]
to_drop = list(filter(lambda x: x, [mask[i] and i for i in range(len(mask))]))
data = data.drop(columns=to_drop)
data = data.merge(act_code, on='Act_Code')
data.Act_Code = data.kind
data.drop(columns=['kind'], inplace=True)
data.groupby(by=['Subject_Code', 'Act_Code']).mean()
```
